# 1. Menghitung Koefisien Regresi dengan Scikit-Learn
Pada metode pertama, saya menggunakan library LinearRegression dari sklearn. Library ini digunakan untuk membuat model regresi linier secara otomatis berdasarkan data X dan Y yang telah disiapkan.

Data X berisi nilai pada sumbu horizontal, sedangkan data Y berisi nilai pada sumbu vertikal.


```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Menyiapkan data titik X dan Y
X = np.array([2, 4, 5, 3, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 4, 3, 5, 6])

# Membuat model regresi linier
model = LinearRegression()
model.fit(X, y)

# Menampilkan hasil koefisien dan intercept
print(f"Nilai Intercept (a)  : {model.intercept_:.4f}")
print(f"Nilai Koefisien (b)  : {model.coef_[0]:.4f}")
```

Berdasarkan hasil tersebut, persamaan garis regresi linier yang diperoleh adalah:
$$
y = 1.0769x + 0
$$

Karena nilai intercept bernilai 0, maka persamaan dapat ditulis menjadi:
$$
y = 1.0769x
$$
